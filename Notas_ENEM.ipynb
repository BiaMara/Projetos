Aula 4 - Machine Learning

Um bom modelo precisa pegar diversas características para poder projetar o valor do que se deseja. NO CASO AQUI, as características são as notas.

```
provas_entrada = ["NU_NOTA_CH", "NU_NOTA_LC", "NU_NOTA_CN", "NU_NOTA_REDACAO"]
prova_saida = "NU_NOTA_MT"
#Mas se tiver NaN, o modelo dá erro, então precisa criar uma base sem esses valores.
#Tirando os NaN da amostra
dados_sem_nota_zero = dados_sem_nota_zero[provas].dropna()
#Criando 2 dataframes diferentes, um para as notas de entrada e um para a nota de saída.
notas_entrada =  dados_sem_nota_zero[provas_entrada]
notas_saida = dados_sem_nota_zero[prova_saida]
```

```
notas_entrada
```

Entrada e saída lembram uma função. f(x) = y

```
x = notas_entrada
y = notas_saida
```

É preciso separar parte da amostra para fazer testes para validar o modelo.
Então divide entre dados de teste e dados de treino.
Treino: para ensinar o modelo.
Teste: pra rodar de fato.
Biblioteca sckitch learn é usada para rodar modelos e regressões.

```
from sklearn.model_selection import train_test_split

x_treino, y_treino, x_teste, y_teste = train_test_split(x, y, test_size = 0.25)
#Estou separando 25% para teste (test_size)
#A divisão de dados de treino e dados de teste são aleatórias.
#Cada vez que rodar, dará um resultado.
```

```
#Setando valor para ter mais folds: cv=10, tenho 10 folds
modelo_arvore = DecisionTreeRegressor(max_depth = 3)
resultados = cross_validate(modelo_arvore, x, y, cv = 10, scoring="neg_mean_squared_error")
resultados["test_score"]*-1
```

```
#Calculando o intervalo de confiança pro modelo
desvio_padrao = (resultados["test_score"]*-1).std()
desvio_padrao
#A média - 2XDP  e media + 2DP: 95% das vezes a amostra estará nesse intervalo
lim_inferior = media - (2*desvio_padrao)
lim_superior = media + (2*desvio_padrao)
print(f"Intervalo de confiança {lim_inferior} - {lim_superior}")
```

```
#Criando uma função
def calcula_mse(resultados): 
  media = (resultados["test_score"]*-1).mean()
  desvio_padrao = (resultados["test_score"]*-1).std()
  lim_inferior = media - (2*desvio_padrao)
  lim_superior = media + (2*desvio_padrao)
  print(f"Intervalo de confiança {lim_inferior} - {lim_superior}")
```

```
modelo_arvore = DecisionTreeRegressor(max_depth = 3)
resultados = cross_validate(modelo_arvore, x, y, cv = 10, scoring="neg_mean_squared_error")
calcula_mse(resultados)
```

```
#Embaralhando os dados para não ficarem viesados
from sklearn.model_selection import  KFold
#Seta o shuffle para embaralhar a amostra
partes = KFold(n_splits = 10, shuffle=True)
modelo_arvore = DecisionTreeRegressor(max_depth = 3)
resultados = cross_validate(modelo_arvore, x, y, cv = partes, scoring="neg_mean_squared_error")
calcula_mse(resultados) 
```

```
#Se precisar de um valor contínuo tem que setar o random
import numpy as np
SEED = 1232
np.random.seed(SEED)
partes = KFold(n_splits = 10, shuffle=True)
modelo_arvore = DecisionTreeRegressor(max_depth = 3)
resultados = cross_validate(modelo_arvore, x, y, cv = partes, scoring="neg_mean_squared_error")
calcula_mse(resultados) 
```

```
#Hiperparâmetro: alteram o resultados da função
#Quanto mais ramos na árvore, melhor pode ser o resultado.
#Criando uma função para rodar com diversos niveis de arvore
def regressor_arvore(nivel): 
    SEED = 1232
    np.random.seed(SEED)
    partes = KFold(n_splits = 10, shuffle=True)
    modelo_arvore = DecisionTreeRegressor(max_depth = nivel)
    resultados = cross_validate(modelo_arvore, x, y, cv = partes, scoring="neg_mean_squared_error")
    print(f"Teste = {(resultados['test_score']*-1).mean()}")

regressor_arvore(4)

```

```
#Variando a profundidade de 1 até 20
for i in range(1,21):
    regressor_arvore(i)

#Melhora até um certo porto, depois o mse volta a crescer. 
#Ele vai bem até o 8º nivel mais ou menos
```
